# Bifurcaciones con el parámetro $\tilde{Q}$

```python
from model import *
from scipy.signal import find_peaks
```

```python
m = {'1': 0.125, '3': 0.005} # g
d = {'1': 0.25, '2': 0.05, '3': 0.01} # cm
r = {'1': 0.01, '2': 0.005, '3': 0.005} # g/ms
k = {'1': 0.08, '3': 0.01, '1,2': 0.05, '2,3': 0.01} # g/ms^2
c = {'1': 3 * k['1'], '3': 3 * k['3']} # g/ms^2
a0 = {'1': 0.01, '2': 0.01, '3': 0.01} # cm^2
l = 1.4 # cm
rho = 0.00113 # g/cm^3
Ps = 0.008 # g/cm ms^2
k20 = 0.08 # g/ms^2
m20 = 0.025 # g
```

```python
y0 = [0.1, 0.1, 0.1, 0, 0, 0]
rate = 20000
t = np.linspace(0, 1000, rate + 1)
```

```python
def get_bifurcation_values(values, init, Q_max, l_max_list, Q_min, l_min_list, Q_freq, freq_list):
    total = {'x$_1$': [], 'x$_2$': [], 'x$_3$': [], 'U': []}
    
    for val in values:
        k['2'] = val * k20
        c['2'] = 3 * k['2']
        m['2'] = m20 / val

        sol = odeint(model, init, t, args=(m, d, r, k, c, a0, l, Ps))
        x1, x2, x3 = sol[:, 0], sol[:, 1], sol[:, 2]

        a1 = a0['1'] + 2 * l * x1
        a2 = a0['2'] + 2 * l * x2
        a3 = a0['3'] + 2 * l * x3

        a_min = np.asarray([a1, a2, a3]).min(0)
        U = np.sqrt((2 * Ps) / rho) * a_min * np.vectorize(theta)(a_min)
        
        signals = {'x$_1$': x1, 'x$_2$': x2, 'x$_3$': x3, 'U': U}
        
        for name, signal in signals.items():
            # Calculo los últimos 20 máximos locales
            peaks_max, _ = find_peaks(signal)
            l_max = signal[peaks_max][-20:]
            Q_max[name] += [val] * l_max.size
            l_max_list[name] += list(l_max)
            
            # Calculo los últimos 20 mínimos locales
            peaks_min, _ = find_peaks(-1 * signal)
            l_min = signal[peaks_min][-20:]
            Q_min[name] += [val] * l_min.size
            l_min_list[name] += list(l_min)
            
            # Calculo la frecuencia
            fourier = np.fft.fft(signal[-7000:] - signal[-7000:].mean())
            freq = np.fft.fftfreq(fourier.size) * rate
            fourier = np.abs(fourier[0:fourier.size // 2])
            freq = freq[0:freq.size // 2]
            freq_max = 0 if np.max(fourier) < 1 else np.argmax(fourier)
            freq_list[name].append(freq[freq_max])
            Q_freq[name].append(val)
            
            # Agrego la señal al total
            total[name] += list(signal[-8192:] - signal[-8192:].mean())
        
        init = sol[-1,:]
    return sol[-1,:], total
```

```python
Q_max = {'x$_1$': [], 'x$_2$': [], 'x$_3$': [], 'U': []}
Q_max_rev = {'x$_1$': [], 'x$_2$': [], 'x$_3$': [], 'U': []}
local_max = {'x$_1$': [], 'x$_2$': [], 'x$_3$': [], 'U': []}
local_max_rev = {'x$_1$': [], 'x$_2$': [], 'x$_3$': [], 'U': []}
Q_min = {'x$_1$': [], 'x$_2$': [], 'x$_3$': [], 'U': []}
Q_min_rev = {'x$_1$': [], 'x$_2$': [], 'x$_3$': [], 'U': []}
local_min = {'x$_1$': [], 'x$_2$': [], 'x$_3$': [], 'U': []}
local_min_rev = {'x$_1$': [], 'x$_2$': [], 'x$_3$': [], 'U': []}
Q_freq = {'x$_1$': [], 'x$_2$': [], 'x$_3$': [], 'U': []}
Q_freq_rev = {'x$_1$': [], 'x$_2$': [], 'x$_3$': [], 'U': []}
freq = {'x$_1$': [], 'x$_2$': [], 'x$_3$': [], 'U': []}
freq_rev = {'x$_1$': [], 'x$_2$': [], 'x$_3$': [], 'U': []}
Q_values = np.linspace(0.5, 1.8, 81)

last, total = get_bifurcation_values(Q_values, y0, Q_max, local_max, Q_min, local_min, Q_freq, freq)
_, total_rev = get_bifurcation_values(Q_values[::-1], last, Q_max_rev, local_max_rev, \
                                      Q_min_rev, local_min_rev, Q_freq_rev, freq_rev)
```

```python
for name in Q_max:
    plt.figure(figsize=(10,5))
    plt.title('Diagrama de bifurcación')
    plt.xlabel(r'Parámetro de Tensión $\tilde{Q}$')
    plt.ylabel('Máximo Local de {} [{}]'.format(name, 'cm' if name != 'U' else 'cm$^3$/ms'))
    plt.scatter(Q_max[name], local_max[name], facecolors='none', edgecolors='C0', s=64)
    plt.scatter(Q_max_rev[name], local_max_rev[name], marker='x', s=64)
    plt.show()
```
